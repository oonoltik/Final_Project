Объединяем в один дата фрейм файлы из csse_covid_19_daily_reports

```python
def join_several_to_one (absolute_path, path, result_file_name, condition):  
    #делаем список имен файлов для единого датасета с исключением ненужных лет в названии
    abs_list = []
    for root, dirs, files in os.walk(path):  
        for filename in files:
            if condition in filename and '.csv' in filename:
#                 p = absolute_path + path.replace('.\\','') +'\\'+  filename
                p = absolute_path + path.replace('.\\','').replace('Start_datasets_COVID\\','') + '\\' + filename
#                 print(p)             
                abs_list.append(p)
            else:
    #             print('Нашел 2022', filename)
                continue
   
    print(len(abs_list))
    
    #делаем единый датасет
    frame = pd.DataFrame()
    list_ = []
    for file_ in abs_list:
        df = pd.read_csv(file_,index_col=None, header=0)
        list_.append(df)
    frame = pd.concat(list_)
    
    #Видим, что поле дата все равно папопали даты содержащие 2022 год убираем их, оставляя в поле только 2021 год
    frame = frame[frame['Last_Update'].apply(lambda x: condition in x)]
    
    frame.to_csv(result_file_name, index=False)
    
    return frame
```
