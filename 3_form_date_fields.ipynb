# Работаем с полем дата в датасетах - приводим все к унифицированной форме для дальнейшей возможности сравнить данные из разных датасетов

```python
import pandas as pd
import datetime
import time
```

```python
PREPARED_DATASET_PATH = 'joined_dataset_pollution.csv'
PREPARED_DATASET_PATH_2 = 'joined_dataset_COVID_1.csv'
```

Сначала Работаем с joined_dataset_pollution

```python
df = pd.read_csv(PREPARED_DATASET_PATH)
df = df.sort_values(by='Date')
df
```

Смотрим уникальные значения по полю Дата, чтобы убедится, что все даты представлены в одном формате

```python
dates_unique_1 = df.Date.unique()
print(dates_unique_1)
print(len(dates_unique_1))
```

```python
df.info()
```

```python
# функция преобразования даты в кол-во секунд с начала эпохи (абсолютное время)
def abs_time(date):
#     print(date)
    date = str(date.replace('-',''))
    abs_time = time.mktime(time.strptime(date, "%Y%m%d"))
    abs_time = int(abs_time)
#     print(abs_time)
    return abs_time
```

```python
# Проверка функции:
date1 = '2021-01-03'
a = abs_time(date1)
a
```

```python
%%time
#Добавляем сттолбец с абсолютным временем в датафрейм

df['Absolute_time'] = df['Date'].apply(abs_time)
df
```

```python
#сохраняем датасет под индексом 2
df.to_csv('joined_dataset_pollution_2.csv', index=False)
```

# Теперь Работаем с joined_dataset_COVID_1

```python
df_2 = pd.read_csv(PREPARED_DATASET_PATH_2)
df_2
```

Меняем название Last_Update на Date

```python
df_2.rename(columns={'Last_Update': 'Date'}, inplace=True)
```

```python
dates_unique_2 = df_2.Date.unique()
print(dates_unique_2)
print(len(dates_unique_2))
```

Отсекаем из поля даты значения времени (часы, минуты, секунды). Оставляем только дату:

```python
%%time
# сначала преобразуем  в to_datetime с отсечением по дате
df_2['Date'] = pd.to_datetime(df_2['Date']).dt.date

# потом поле Дата обратно в формат str иначе не сработает функция абсолютного времени потом
df_2['Date'] = df_2['Date'].astype('str')

df_2 = df_2.sort_values(by='Date')
```

```python
df_2
```

```python
df_2.info()
```

```python
dates_unique_2 = df_2.Date.unique()
print(dates_unique_2)
print(len(dates_unique_2))
```

```python
%%time

#Добавляем сттолбец с абсолютным временем в датафрейм
df_2['Absolute_time'] = df_2['Date'].apply(abs_time)
df_2
```

```python
#сохраняем датасет под индексом 2
df_2.to_csv('joined_dataset_COVID_2.csv', index=False)
```
