```python
# функция добавления колонки прироста по заданой колоке Death, Confirmed, Recovered
def calc_increase (df, col_name):
    df_ura = df
    new_name = "Day_increase" + "_" +  col_name 
    df_ura[new_name] =  ""
    unique_cnytrs = df_ura['Dugramm_COV'].unique()
    for country in unique_cnytrs:
        unique_towns = df_ura.loc[df['Dugramm_COV'] == country]['City_COV'].unique()
        for town in unique_towns:            
#             df.loc[(df['Dugramm_COV'] == country) & (df['City_COV']  == town), new_name] =  df.loc[(df['Dugramm_COV'] == country) & (df['City_COV']  == town)][col_name] - df.loc[(df['Dugramm_COV'] == country) & (df['City_COV']  == town)][col_name].shift(periods=1)
            
            df_ura.loc[(df_ura['City_COV'] == town) & (df_ura['Dugramm_COV'] == country), new_name] =\
            df_ura.loc[(df_ura['City_COV'] == town) & (df_ura['Dugramm_COV'] == country)][col_name] - \
            df_ura.loc[(df_ura['City_COV'] == town) & (df_ura['Dugramm_COV'] == country)][col_name].shift(periods=1) 
    
    return df_ura   
#  df.loc[df['Dugramm_COV'] == country]['City_COV'] == town, 'Deaths_increase'] =  df.loc[df['Dugramm_COV'] == country]['City_COV'] == town]['Deaths_COV'] - df.loc[df['Dugramm_COV'] == country]['City_COV'] == town]['Deaths_COV'].shift(periods=1) 
   
#     print(unique_towns)
#     df.loc[df['City_COV'] == 'Athens', 'Deaths_increase'] =  df_ura.loc[df_ura['City_COV'] == 'Athens']['Deaths_COV'] - df_ura.loc[df_ura['City_COV'] == 'Athens']['Deaths_COV'].shift(periods=1) 
```

```python
# добавляем колонки прироста по  колонкам Death, Confirmed, Recovered:
col_list = ['Deaths_COV', 'Recovered_COV', 'Active_COV']
for col in col_list:
    df_2 = calc_increase(df_2, col)
df_2
```
